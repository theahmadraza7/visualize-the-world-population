```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the files
population_data = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_2001050.csv', skiprows=4)
metadata_country = pd.read_csv('Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_2001050.csv')
metadata_indicator = pd.read_csv('Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_2001050.csv')
```

```python
# Extract population data for 2023
population_2023 = population_data[['Country Name', '2023']].dropna()
population_2023.columns = ['Country', 'Population']
```

```python
# Sort by population and select the top 10 countries
top10_population_2023 = population_2023.sort_values(by='Population', ascending=False).head(10)
```

```python
# Create a bar chart
plt.figure(figsize=(12, 8))
plt.bar(top10_population_2023['Country'], top10_population_2023['Population'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Most Populous Countries in 2023')
plt.xticks(rotation=45)
plt.show()
```
